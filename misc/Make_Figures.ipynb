# 1. Influence of alpha in softmin

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
```

```python

def param_softmin(x, alpha=1.0):
    exponentials = np.exp(-alpha*x)
    return exponentials / np.sum(exponentials)
```

```python
np.random.seed(42)
x = np.abs(np.random.random(10))
x = x.astype('float64')
x
```

```python
x = np.array([0.5, 0.1, 0.7, 0.8, 0.4, 0.3], dtype='float64')
```

```python
alphas = [1.0, 5.0, 10.0, 20.0, 50.0, 100.0]
probs = np.array([param_softmin(x, alpha=a) for a in alphas])
```

```python
probs.sum(axis=1)
```

```python
plt.plot(x, label="Original values", linestyle='dashed')
for p, a in zip(probs, alphas):
    plt.plot(p, label=r'Softmin with $\alpha$=' + str(a))
plt.legend(title=r'Influence of $\alpha$ in softmin function', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.savefig('alpha_softmin.png', bbox_inches='tight', dpi=400)
plt.show()
```
